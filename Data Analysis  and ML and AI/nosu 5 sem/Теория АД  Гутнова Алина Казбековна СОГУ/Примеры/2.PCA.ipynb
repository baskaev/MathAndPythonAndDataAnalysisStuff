```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
y = 2 * x + np.random.randn(10)*2
X = np.vstack((x,y))
print(X)
```

```python
plt.scatter(x, y);
```

```python
Xcentered = (X[0] - x.mean(), X[1] - y.mean())
m = (x.mean(), y.mean())
print(Xcentered)
print("Mean vector: ", m)
```

```python
plt.scatter(Xcentered[0], Xcentered[1]);
plt.scatter([0], [0], c="green");
```

```python
x = np.arange(1,101)
y = -1.3 * x + np.random.randn(100)*10
X = np.vstack((x,y))
print(X)
```

```python
Xcentered = (X[0] - x.mean(), X[1] - y.mean())
m = (x.mean(), y.mean())
print(Xcentered)
print("Mean vector: ", m)
```

```python
plt.scatter(Xcentered[0], Xcentered[1]);
plt.scatter([0], [0], c="green");
```

```python
covmat = np.cov(Xcentered)
print(covmat, "\n")
print("Variance of X: ", covmat[0,0])
print("Variance of Y: ", covmat[1,1])
print("Covariance X and Y: ", covmat[0,1])
```

```python
covmat = np.cov(Xcentered)
print(covmat, "\n")
print("Variance of X: ", covmat[0,0])
print("Variance of Y: ", covmat[1,1])
print("Covariance X and Y: ", covmat[0,1])
```

```python
val, vecs = np.linalg.eig(covmat)
print(val, vecs)
v = -vecs[:,1]
Xnew = np.dot(v,Xcentered)
print(Xnew)
```

```python
n = 9 #номер элемента случайной величины
Xrestored = np.dot(Xnew[n],v) + m
print('Restored: ', Xrestored)
print('Original: ', X[:,n])
```
