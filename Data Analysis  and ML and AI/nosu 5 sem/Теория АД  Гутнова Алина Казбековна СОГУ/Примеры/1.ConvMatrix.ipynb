```python
from PIL import Image
import numpy as np
import time

start_time = time.time()
img = Image.open("Lenna.png").convert('L')
m_img = img.load()
kernel = [[0.000789,  0.006581, 0.013347, 0.006581, 0.000789],
[0.006581,  0.054901, 0.111345, 0.054901, 0.006581],
[0.013347,  0.111345, 0.225821, 0.111345, 0.013347],
[0.006581,  0.054901, 0.111345, 0.054901, 0.006581],
[0.000789,  0.006581, 0.013347, 0.006581, 0.000789]]

w, h = img.size
padding = len(kernel) // 2
data = [[-1] * (w - 2 * padding) for _ in range(h - 2 * padding)]

for i in range(padding, w - padding):
    for j in range(padding, h - padding):
        sum_ = 0
        for k in range(-padding, padding + 1):
            for l in range(-padding, padding + 1):
                sum_ += m_img[i + k, j + l] * kernel[k + padding][l + padding]
        data[i - padding][j - padding] = sum_

new_img = Image.fromarray(np.array(data).astype(np.uint8).T)
new_img.save("blur_b.png")
print(time.time() - start_time)
```
